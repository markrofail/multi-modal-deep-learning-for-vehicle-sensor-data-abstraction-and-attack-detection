# Resizing Images

```bash
original_size = 1242x375
new_size      = 1392x512    # <-- size used in Regnet
```

## 1.resize (upsample)

```bash
convert -resize 1392x574 -interpolate bilinear '0000000000-original.png ' '0000000000-resize1.png'
convert -resize 1696x512 -interpolate bilinear '0000000000-original.png ' '0000000000-resize2.png'
```

## 2.crop

```bash
convert -gravity center -crop 1392x512+0+0 '0000000000-resize1.png' '0000000000-crop1.png'
convert -gravity center -crop 1392x512+0+0 '0000000000-resize2.png' '0000000000-crop2.png'
```

## 3.entropy

```bash
convert '0000000000-original.png' -print '%[entropy]\n' null:
```

original_ent = 0.888018

```bash
convert '0000000000-resize1.png' -print '%[entropy]\n' null:
```

resize1_ent = 0.890994

```bash
convert '0000000000-resize2.png' -print '%[entropy]\n' null:
```

resize2_ent = 0.890605

```bash
convert '0000000000-crop1.png' -print '%[entropy]\n' null:
```

crop1_ent = 0.890994

```bash
convert '0000000000-crop2.png' -print '%[entropy]\n' null:
```

crop2_ent = 0.894234

## 4.information gain
Information gain = before - after

```bash
inf_diff_crop1 = 0.888018 - 0.890994 = −0.002976
inf_diff_crop2 = 0.888018 - 0.894234 = −0.006216
```

thus, we'll go with the crop 1 settings as the information loss is lower
